```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

 1. Prepare a list of offices sorted by country, state, city.

```python
run("SELECT * FROM offices ORDER BY country, state, city")
```

2. How many employees are there in the company?

```python
run("SELECT count(*) as NumberofEmployees FROM employees ")
```

3. What is the total of payments received?

```python
run("SELECT sum(amount) as totalPayments FROM payments")
```

4. List the product lines that contain 'Cars'.

```python
run("SELECT productLine FROM productlines WHERE productline LIKE '%Cars%'")
```

5. Report total payments for October 28, 2004.

```python
run("SELECT amount FROM payments WHERE paymentDate='2004-10-28'")
```

6. Report those payments greater than \$100,000.

```python
run("SELECT customerNumber,checkNumber,paymentDate,amount FROM payments WHERE amount>100000" )
```

7. List the products in each product line.

```python
run("SELECT productLine, productName FROM products ORDER BY productLine")
```

8. How many products in each product line?

```python
run("SELECT productLine,COUNT(*) AS NumberOfProducts FROM products GROUP BY productLine")
```

9. What is the minimum payment received?

```python
run("SELECT MIN(amount) AS MinimumPayment FROM payments")
```

10. List all payments greater than twice the average payment.

```python
run("SELECT * FROM payments WHERE amount > (SELECT AVG(amount) AS AveragePayment FROM payments)")
```

11. What is the average percentage markup of the MSRP on buyPrice?

```python
run("SELECT AVG((MSRP/buyPrice-1)*100) AS AverageMarkupInPercent FROM products")
```

12. How many distinct products does ClassicModels sell?

```python
run("SELECT count(DISTINCT productName) FROM products")
```

13. Report the name and city of customers who don't have sales representatives?

```python
run("SELECT customerName,city from customers WHERE salesRepEmployeeNumber is NULL")
```

14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
run("SELECT concat (firstName,' ',lastName) AS fullName, jobTitle FROM employees WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE'%Manager%'")
```

15. Which orders have a value greater than $5,000?

```python
run("SELECT orderNumber, SUM((quantityOrdered * priceEach)) AS TotalOrderValue FROM orderdetails GROUP BY orderNumber HAVING TotalOrderValue > 5000")
```
