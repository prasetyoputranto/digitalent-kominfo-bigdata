# One to many relationship

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

1.Report the account representative for each customer.

```python
run("""
SELECT 
    customerName,
    CONCAT(firstName," ",lastName) AS accountRepresentative
FROM
    customers a
    INNER JOIN
    employees b 
        ON a.salesRepEmployeeNumber = b.employeeNumber
ORDER BY customerName
""")
```

2.Report total payments for Atelier graphique.

```python
run("""
SELECT 
    a.customerName, 
    SUM(b.amount) AS totalPayment
FROM 
    customers a
    INNER JOIN
    payments b 
        ON a.customerNumber = b.customerNumber
WHERE a.`customerName`='Atelier graphique'
GROUP BY a.customerName
""")
```

3.Report the total payments by date

```python
run("""
SELECT 
    paymentDate,
    SUM(amount) AS totalPayment
FROM
    payments
GROUP BY paymentDate
""")
```

4.Report the products that have not been sold

```python
run("""
SELECT 
    productName,
    quantityOrdered
FROM
    orderdetails a
    RIGHT JOIN
    products b 
        ON a.productCode = b.productCode
WHERE 
    a.orderNumber IS NULL
""")
```

5.List the amount paid by each customer.

```python
run("""
SELECT 
    customerName,
    SUM(amount) AS paidAmount
FROM
    customers a
    INNER JOIN
    payments b 
        ON a.customerNumber = b.customerNumber
GROUP BY customerName
""")
```

6.How many orders have been placed by Herkku Gifts?

```python
run("""
SELECT 
    customerName,
    SUM(orderNumber) AS orderAmount
FROM
    customers a
    INNER JOIN
    orders b 
        ON a.customerNumber = b.customerNumber
WHERE
    customerName='Herkku Gifts'
""")
```

7.Who are the employees in Boston?

```python
run("""
SELECT 
    city,
     CONCAT(firstName," ",lastName) AS employeeName
FROM
    employees a
    INNER JOIN
    offices b 
        ON a.officeCode = b.officeCode
WHERE
    city='Boston'
""")
```

8.Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
SELECT 
    customerName,
    amount
FROM
    customers a
    INNER JOIN
    payments b 
        ON a.customerNumber = b.customerNumber
WHERE amount>100000
ORDER BY amount DESC
""")
```

9.List the value of 'On Hold' orders

```python
run("""
SELECT *
FROM 
    orders
WHERE 
    status='On Hold'
""")
```

10.Report the number of orders 'On Hold' for each customer.

```python
run("""
SELECT 
    customerName,
    COUNT(status) AS status_On_Hold
FROM
    customers a
    INNER JOIN
    orders b 
        ON a.customerNumber = b.customerNumber
WHERE 
    status='On Hold'
GROUP BY 
    customerName
""")
```

# Many to many relationship

1.List products sold by order date.

```python
run("""
SELECT 
    productName,    
    orderDate
FROM
    ((orderdetails a
    INNER JOIN
    orders b 
        ON a.orderNumber = b.orderNumber)
    INNER JOIN
    products c 
        ON a.productCode = c.productCode)  
ORDER BY orderDate 
""")
```

2.List the order dates in descending order for orders for the 1940 Ford Pickup Truck

```python
run("""
SELECT 
    productName,
    orderDate
FROM
    ((orderdetails a
    INNER JOIN
    orders b 
        ON a.orderNumber = b.orderNumber)
    INNER JOIN
    products c 
        ON a.productCode = c.productCode)  
WHERE 
    productName='1940 Ford Pickup Truck'
ORDER BY orderDate DESC 
""")
```

3.List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
SELECT 
    customerName,
    orderNumber,
    amount
FROM
    ((customers a
    INNER JOIN
    orders b 
        ON a.customerNumber = b.customerNumber)
    INNER JOIN
    payments c 
        ON a.customerNumber = c.customerNumber)  
 WHERE amount>25000
""")
```

4.Are there any products that appear on all orders?

```python
run("""
SELECT 
    productCode,
    productName,
    COUNT(productCode) AS totalAppearance
FROM 
    orderdetails 
        INNER JOIN
    products USING(productCode)
GROUP BY 
    productCode 
HAVING 
    totalAppearance=(SELECT COUNT(*) from orders);
""")
```

5.List the names of products sold at less than 80% of the MSRP

```python
run("""
SELECT 
    productName,
    MSRP,
    priceEach
FROM
    products a
    INNER JOIN
    orderdetails b 
        ON a.productCode = b.productCode
WHERE
    priceEach<(0.8*MSRP)
ORDER BY MSRP
""")
```

6.Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
SELECT 
    orderNUmber,
    productName,    
    buyPrice,
    priceEach
FROM
    products a
    INNER JOIN
    orderdetails b 
        ON a.productCode = b.productCode
WHERE
    priceEach>=(2*buyPrice)
ORDER BY priceEach
""")
```

7.List the products ordered on a Monday

```python
run("""
SELECT 
    productName,
    DAYNAME(orderDate) AS Day,
    orderDate
FROM
    ((orderdetails t1
    INNER JOIN
    products t2 
        ON t1.productCode = t2.productCode)
    INNER JOIN
    orders t3 
        ON t1.orderNumber = t3.orderNumber)  
WHERE 
    DAYNAME(orderDate)='Monday'
""")
```

8.What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT 
    b.orderNumber,
    b.productCode,
    productName,
    status,
    quantityInStock AS quantityOnHand
FROM
    ((orders a
    INNER JOIN
     orderDetails b
        ON a.orderNumber = b.orderNumber)
    INNER JOIN
    products c
        ON b.productCode = c.productCode)  
WHERE 
    status='On Hold'
GROUP BY
    productName
ORDER BY
    orderNumber
""")

```
